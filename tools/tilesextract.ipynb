```python
import pandas as pd
import os
tiles = pd.read_csv(os.path.join('..','maps','training.csv'),header=None)
tiles.head()
```

```python
patterns = {}

for index, row in tiles.iterrows():
    nbhr = ''.join(row[0:-1])
    label = row[8]
    if nbhr in patterns:
        if patterns[nbhr] != label:
            print('Inconsistent pattern')
    else:
        patterns[nbhr] = label
        
for pattern in patterns:
    print(pattern,patterns[pattern])
    
```

```python
import json
with open('patterns.json', 'w') as fp:
    json.dump(patterns, fp)
```

```python
with open('patterns.csv', 'w') as f: 
    for pattern in patterns:
        f.write(pattern+','+patterns[pattern]+'\n')
```
